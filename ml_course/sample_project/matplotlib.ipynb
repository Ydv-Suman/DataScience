## complete matplotlib workflow

```python
## step 0 import the library
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

## step 1 enter data
x = np.array([5, 7, 10, 14])
y = np.array([5,10,15,20])

## step 2 setup plot
fig, ax = plt.subplots(figsize=(5,5))  ## figure size = (width, height)

## step 3 plot data
ax.plot(x,y)
## step 4 customize plot
ax.set(title="Title of plot",
       xlabel="x-axis",
       ylabel="y-axis"
)

## step 5 save and show
fig.savefig("images/simple-matplotlib.png")
plt.show()
```

## Different types of plot

```python
x3 = np.linspace(0,20,50)
x3
```

```python
y3 = np.linspace(0,10,50)
y3
```

```python
## line plot
fig, ax = plt.subplots()
ax.plot(x3,x3**2)
plt.show()
```

```python
## scatter plot
fig, ax = plt.subplots()
ax.scatter(x3,np.exp(y3))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x3,np.sin(y3))
plt.show()
```

```python
## Bar plot
fig, ax = plt.subplots()
ax.bar(x3[:10],y3[:10])
plt.show()
```

```python
## making bar plot using dictionary
weekly_sales = {
    "physics": 20,
    "Chemistry": 18,
    "Math": 15
}
fig,ax = plt.subplots()
ax.bar(weekly_sales.keys(), weekly_sales.values())
ax.set(
    title="Stationery Weekly sales report",
    xlabel="List of Books",
    ylabel="Number of books sold"
)
plt.show()
```

```python
## making bar plot using dictionary
weekly_sales = {
    "physics": 20,
    "Chemistry": 18,
    "Math": 15
}
fig,ax = plt.subplots()
ax.barh(weekly_sales.keys(), weekly_sales.values())   ## barh plot the graph horizontally
ax.set(
    title="Stationery Weekly sales report",
    xlabel="List of Books",
    ylabel="Number of books sold"
)
plt.show()
```

```python
## Histogram
x5= np.random.randint(0,10)
fig, ax = plt.subplots()
ax.hist(x5)
plt.show()
```

```python
## subplots option 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10,8)
) ## must provide size of row and columns

## ploting data for each axes
ax1.plot(x3,np.sin(y3))
ax2.scatter(x3,np.sin(y3))
ax3.bar(x3[:10],np.sin(y3[:10]))
ax4.hist(x5)



plt.show()

```

```python
## subplots options 2
fig, ax = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10,5)
)
ax[0,0].plot(x3,np.sin(y3))
ax[0,1].scatter(x3, np.sin(y3))
ax[1,0].barh(x3,y3)
ax[1,1].hist(np.random.randn(1000))
plt.show()
```

## plotting from pandas dataframes

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
## first replace $,. to perform calculations
## using str[:-2], there are two extra zeros 
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,.]', '', regex=True).str[:-2]
car_sales
```

```python
car_sales["Date"] = pd.date_range(start='05/06/2025', periods=len(car_sales))
car_sales
```

```python
date_col = car_sales.pop("Date")
car_sales.insert(0, "Date", date_col)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)     ## using astype int, price is str
car_sales["Total"] = car_sales["Price"].cumsum()  
car_sales
```

```python
## ploting the total sales
car_sales.plot(x="Date", y="Total")
plt.show()
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
plt.show()
```

```python
car_sales.plot.barh(x="Date", y="Total")
plt.show()
```

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].astype(int)
car_sales["Odometer (KM)"].plot.hist(bins=10)    ## the bins parameter controls how the data is grouped into intervals
                                                ## bins=5  divides the data into 5 ranges
plt.show()
```

```python
car_sales.plot.hist(subplots=True)
plt.show()
```
