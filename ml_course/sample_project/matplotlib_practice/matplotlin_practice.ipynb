```python
## from dictionary
## using OO method
exercise = {
    "PushUp": 10,
    "PullUp": 8,
    "Plank": 4,
    "Running": 20
}

fig, ((ax1, ax2),(ax3, ax4),(ax5,ax6)) = plt.subplots(
    nrows=3,
    ncols=2,
    figsize=(18,20)
)

## for ax1
ax1.scatter(exercise.keys(), exercise.values())
ax1.set(
    title="scatter",
    xlabel="Types of Exercise",
    ylabel= "Minutes"
)

## for ax2
ax2.plot(exercise.keys(), exercise.values())
ax2.set(
    title="plot ",
    xlabel="Types of Exercise",
    ylabel= "Minutes"
)

## for ax3
ax3.bar(exercise.keys(), exercise.values())
ax3.set(
    title="bargraph",
    xlabel="Types of Exercise",
    ylabel= "Minutes"
)

## for ax4
ax4.hist(exercise.values(), bins=(3))
ax4.set(
    title="histogram",
    xlabel="Types of Exercise",
    ylabel= "Minutes"
)
## for ax5
ax5.plot(exercise.keys(), exercise.values())
ax5.scatter(exercise.keys(), exercise.values())
ax5.bar(exercise.keys(), exercise.values())
ax5.set(
    title="Mixed",
    xlabel="Types of Exercise",
    ylabel= "Minutes"
)

fig.suptitle("Exercise", fontsize="40", fontweight="bold" )
plt.show()
```

```python
## making dataframe
students = pd.DataFrame({
    "Name": ["Daniel", "Sophia", "Liam", "Emma", "Noah", "Olivia", "James", "Ava", "Benjamin", "Mia"],
    "Attendance": [80, 92, 88, 95, 77, 85, 90, 78, 83, 89],
    "GPA": [3.8, 3.6, 3.2, 3.9, 2.8, 3.5, 3.7, 3.0, 3.4, 3.6],
    "Donation": ["$30.00", "$50.00", "$20.00", "$35.00", "$15.00", "$25.00", "$40.00", "$10.00", "$28.00", "$32.00"]
})
students
```

```python
students.to_csv("../matplotlib_practice/students.csv", index=False) ## index= false to prevent addition of index column
```

```python
student_details = pd.read_csv("students.csv")
student_details
```

```python
student_details["GPA"].plot()
plt.show()
```

```python
student_details["Donation"] = student_details["Donation"].str.replace(r'[\$,.]', '', regex=True).str[:-2].astype(int)
student_details
```

```python

student_details["TotalDonation"] = student_details["Donation"].cumsum()
student_details
```

```python
student_details["Date"] = pd.date_range("03/02/2025", periods=len(student_details))
student_details
```

```python
fig, ax = plt.subplots(
    figsize=(20,7)
)
ax.plot(student_details["Date"], student_details["TotalDonation"])
plt.show()
```

```python
fig, ax = plt.subplots(
    figsize=(20,7)
)
ax.scatter(student_details["GPA"], student_details["TotalDonation"])
plt.show()
```

```python
## using style
plt.style.use('seaborn-v0_8-darkgrid')

fig, ((ax1, ax2),(ax3, ax4),(ax5, ax6)) = plt.subplots(
    nrows=3,
    ncols=2,
    figsize=(25,25)
)
## for ax1
scatter = ax1.scatter(
    x = kidney_failure['Age'],
    y = kidney_failure['Blood_Pressure'],
    c = kidney_failure['Kidney_Failure'],
    cmap=('winter')
)
ax1.set(
    title = 'Kideney failure and Blood_Pressure',
    xlabel = 'Age',
    ylabel = 'Blood Pressure'
    
)
ax1.set_xlim([10,90])
ax1.legend(*scatter.legend_elements(), title='kidney failure')
ax1.axhline(
    kidney_failure['Blood_Pressure'].mean(),
    linestyle='--'
)

## for ax2
scatter = ax2.scatter(
    x = kidney_failure['Age'],
    y = kidney_failure['Blood_Urea'],
    c = kidney_failure['Kidney_Failure'],
    cmap=('winter')
)
ax2.set(
    title = 'Kideney failure and Blood Urea',
    xlabel = 'Age',
    ylabel = 'Blood_Urea'
    
)
ax2.set_xlim([10,90])
ax2.legend(*scatter.legend_elements(), title='kidney failure')
ax2.axhline(
    kidney_failure['Blood_Urea'].mean(),
    linestyle='--'
)

## for ax3
scatter = ax3.scatter(
    x = kidney_failure['Age'],
    y = kidney_failure['Serum_Creatinine'],
    c = kidney_failure['Kidney_Failure'],
    cmap=('winter')
)
ax3.set(
    title = 'Kideney failure and Serum_Creatinine',
    xlabel = 'Age',
    ylabel = 'Serum Creatinine'
    
)
ax3.set_xlim([10,90])
ax3.legend(*scatter.legend_elements(), title='kidney failure')
ax3.axhline(
    kidney_failure['Serum_Creatinine'].mean(),
    linestyle='--'
)

## for ax4
scatter = ax4.scatter(
    x = kidney_failure['Age'],
    y = kidney_failure['Hemoglobin'],
    c = kidney_failure['Kidney_Failure'],
    cmap=('winter')
)
ax4.set(
    title = 'Kideney failure and Hemoglobin',
    xlabel = 'Age',
    ylabel = 'Hemoglobin'
    
)
ax4.set_xlim([10,90])
ax4.legend(*scatter.legend_elements(), title='kidney failure')
ax4.axhline(
    kidney_failure['Hemoglobin'].mean(),
    linestyle='--'
)

## for ax5
scatter = ax5.scatter(
    x = kidney_failure['Age'],
    y = kidney_failure['Sodium'],
    c = kidney_failure['Kidney_Failure'],
    cmap=('winter')
)
ax5.set(
    title = 'Kideney failure and Sodium',
    xlabel = 'Age',
    ylabel = 'Sodium'
    
)
ax5.set_xlim([10,90])
ax5.legend(*scatter.legend_elements(), title='kidney failure')
ax5.axhline(
    kidney_failure['Sodium'].mean(),
    linestyle='--'
)


plt.show()
fig.savefig("../matplotlib_practice/kidneyFailure.png")
```
